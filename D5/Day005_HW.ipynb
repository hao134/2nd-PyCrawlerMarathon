```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
```

```python
import json
data = json.loads(response)
data
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

print(len(data))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

# popular
total_like = 0
total_comment = 0
for d in data:
    total_like += d['likeCount']
    total_comment += d['commentCount']
print('Average like number:',total_like/len(data))
print('Average comment number:',total_comment/len(data))
```

```python
# non_popular
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = r.text
data = json.loads(response)

total_like = 0
total_comment = 0
for d in data:
    total_like += d['likeCount']
    total_comment += d['commentCount']
print('Average like number:',total_like/len(data))
print('Average comment number:',total_comment/len(data))
```
